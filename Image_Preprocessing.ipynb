```python
import cv2 , numpy as np
from google.colab.patches import cv2_imshow
#Read image
img=cv2.imread('img.jpeg')
```

```python
#Creating 5x5 matrix
k = np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]], np.float32)
k
```

```python
k = k / k.sum() #Normalize kernel
#Applying blur using custom kernel
blur=cv2.filter2D(img,-1,k)

#Apply Canny Edge Detection
edges=cv2.Canny(blur,50,150)

#show image
print("Original image ")
cv2_imshow(img)
print("Blurred image ")
cv2_imshow(blur)
print("Canny Edges image ")
cv2_imshow(edges)
```

```python
#Sharping the image
kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
sharpened = cv2.filter2D(img, -1, kernel)
print("Original image ")
cv2_imshow(img)
print("Sharpended Image")
cv2_imshow(sharpened)
```
