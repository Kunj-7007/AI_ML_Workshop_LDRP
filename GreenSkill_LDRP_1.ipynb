```python
import pandas as pd
```

**Sample data with values (using None )**

```python

data={
    "Energy source":["Solar","Wind","Hydropower","Geothermal","Biomass","Other"],
    "Energy Consumption (Mwh)":[100,None,150,250,None,500],
    "Cost (Millons)":[200,None,150,250,None,125],
    }
energy_df=pd.DataFrame(data)
print("Original Data with empty values ")
print(energy_df)
```

#**Replace the Nan by droping**

```python

clean_df=energy_df.dropna()
print("Cleaned Data without empty values")
print(clean_df)
```

# Filling missing values using mean


```python
# Fill missing values in 'Energy Consumption (Mwh)' with the mean of the column
energy_df['Energy Consumption (Mwh)'] = energy_df['Energy Consumption (Mwh)'].fillna(energy_df['Energy Consumption (Mwh)'].mean())

# Fill missing values in 'Cost (Millons)' with the mean of the column
energy_df['Cost (Millons)'] = energy_df['Cost (Millons)'].fillna(energy_df['Cost (Millons)'].mean())

print("DataFrame after filling missing values with column means:")
print(energy_df)
```

```python
#Forward filling missing values
#ffill for previous valu and bfill for next value
foreward_filled_df = energy_df.ffill()
print("DataFrame after forward filling missing values:")
print(foreward_filled_df)
```

```python
#bfill for backward values
backward_filled_df = energy_df.bfill()
print("DataFrame after forward filling missing values:")
print(backward_filled_df)
```

##Create a falg coloumn for missing values and cost values


```python
#create a falg coloumn for missing values and cost values
energy_df["Energy_Consumption Missing"]=energy_df["Energy Consumption (Mwh)"].isna()
energy_df["Cost_Missing_Values"]=energy_df["Cost (Millons)"].isna().astype(int)
print(energy_df)
```

# DATA PRE PROCESSING
##**2.1 Normalisation**
    By using Max and Min Sacler

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
energy_df[["Energy Consumption (Mwh)","Cost (Millons)"]]=scaler.fit_transform(energy_df[["Energy Consumption (Mwh)","Cost (Millons)"]])
print("\nData After normalization")
print(energy_df)
```

**2.2 Satndardize (using Z-score Scaling)**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
energy_df[["Energy Consumption (Mwh)","Cost (Millons)"]]=scaler.fit_transform(energy_df[["Energy Consumption (Mwh)","Cost (Millons)"]])
print("\n After using (Z-Score Saclling)")
print(energy_df)
```

**2.3 Encoding Categorial Variables**

We will convert categorial coloumn Energy Scourse into numerical

```python
energy_encoded_df=pd.get_dummies(energy_df,columns=["Energy source"])
energy_encoded_df=energy_encoded_df.astype(int)
print ("After encoding")
print("\n ", energy_encoded_df)
```
