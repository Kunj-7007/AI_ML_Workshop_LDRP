```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input , Conv1D,  GlobalMaxPooling1D ,Dense
#For evaluation metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 7 Days Electricity Usage Pattern
x=np.array([
    [3],[3],[3],[3],[3],[3],[3],
    [2],[2],[2],[2],[2],[2],[2],
    [7],[7],[7],[7],[7],[7],[7],
    [8],[8],[8],[8],[8],[8],[8],
    ], dtype="float32")

y=np.array([1,1,0,0],dtype="float32")
```

```python
model=Sequential([
#Input layer :
#Example: 7 days of data, each day having 1 value
Input (shape=(7, 1)),
#1D Convolution layer:
#4 filters (kernels) are used
#Kernel size 3 (looks at 3 time steps at a time)
#ReLU activation adds non-linearity
Conv1D(4, 3, activation="relu"),
#Global Max Pooling:
#Takes the maximum value from each filter output
#Reduces data size and keeps the most important features
GlobalMaxPooling1D(),
#Dense (output) layer:
#1 neuron because this is binary classification
#Signoid activation gives output between 0 and 1 (probability)
Dense(1, activation="sigmoid")
])

```

```python
test1 = np.array([[[3],[3],[3],[3],[3],[3],[3]]], dtype="float32")
test2 = np.array([[[8],[8],[8],[8],[8],[8],[8]]], dtype="float32")

print("Efficient prob (low usage):", model.predict(test1, verbose=0))
print("Efficient prob (high usage):", model.predict(test2, verbose=0))
```

```python
#For new data
# x_test shape:[(5,4,9,7)]
# y_test shape:[(2,8,5,1)]

x = x.reshape(4, 7, 1,) # Reshape x to match the 4 patterns and 7-day structure
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=42)

print("Train sample :",len(x_train))
print("Test sample :",len(x_test))
```

```python
#Compile with accuracy metric
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.summary()
```

```python
#train
history=model.fit(x_train,y_train,epochs=50,verbose=2)
```

```python
# Step 1 :predict probabilities on test data
y_prob=model.predict(x_test).ravel()
print("Predict probabilities :",y_prob)
#step 2 Convert probabilites into class label
threshold =0.5
y_pred=(y_prob>=threshold).astype(int)
print("Predicted class :",y_pred)
print("True Class :",y_test)
```

```python
# Step 3 : Accuracy
acc=accuracy_score(y_test,y_pred)
print("Accuracy :",acc)
```

```python
# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# classification report
cr = classification_report(y_test, y_pred , target_names=["Not efficient", "Efficient"])
print("Classification Report:")
print(cr)
```
