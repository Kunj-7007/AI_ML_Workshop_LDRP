```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({
    "electricity_kwh":[120,200,130,145,140,150,160,170,180,190],
    "water_liters":[90,95,100,105,115,157,169,141,201,210]
})
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["electricity_kwh"], df["water_liters"])
plt.xlabel("electricity_kwh")
plt.ylabel("water_liters")
plt.title("Raw Data : Household Resourse Usage")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=42,n_init=10)
df["cluster"]=kmeans.fit_predict(df[["electricity_kwh","water_liters"]])
```

```python
print("\n Table: Electricity , Water , Target (Cluster)\n ")
print(df[["electricity_kwh","water_liters","cluster"]].to_string(index=False))
```

```python
#new data clustering (Giving data)
new_household=[[300,400]]
new_cluster=kmeans.predict(new_household)[0]
print("\n New Household belongs to cluster :",new_cluster)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='electricity_kwh', y='water_liters', hue='cluster', data=df, palette='viridis', s=100)
plt.title('Household Resource Usage Clustered Data')
plt.xlabel('Electricity (kWh)')
plt.ylabel('Water (liters)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
