```python
from matplotlib import colors
import matplotlib.pyplot as plt

#Sample data energy consumption over 6 months (in Mwh)
months =['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption = [1100,1200,1300,1500,1400,1600]
colors = ['lightcoral','lightgreen','lightblue','lightyellow','orange','mediumpurple']
```

#Create a line plot

```python

plt.plot(months ,energy_consumption , marker ='o' , color='b', linestyle='--')

#add titles and labels
plt.title ('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')
plt.show
```

#Create a Bar plot

```python
#Create a Bar plot
plt.bar(months ,energy_consumption  , color='orange')

#add titles and labels
plt.title ('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')
plt.show
```

#Create a Pie plot

```python

plt.pie(energy_consumption, labels=months, autopct='%1.1f%%' , colors=colors)

#add titles and labels
plt.title ('Energy Consumption Over 6 Months')
# plt.xlabel('Months') # Not typically used for pie charts
# plt.ylabel('Energy Consumption (Mwh)') # Not typically used for pie charts
plt.show()
```

#Create a Dot plot

```python
#Create a Dot plot
plt.scatter(months, energy_consumption, c="r")

#add titles and labels
plt.title ('Energy Consumption Over 6 Months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (Mwh)')
plt.show()
```

#Create a Customized Bar plot

##

```python
#Create a Customized Bar plot
plt.bar(months ,energy_consumption  , color='orange', edgecolor='black',linewidth=2)

#add titles and labels
plt.title ('Energy Consumption Over 6 Months',fontsize=14,fontweight='bold')
plt.xlabel('Months',fontsize=12)
plt.ylabel('Energy Consumption (Mwh)',fontsize=12)

#Adding grid and customize style
plt.grid(True,linestyle='--',alpha=0.7)
plt.legend(['Energy Consumption'],loc='upper left')
plt.show()
```

## Heat map using seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create a dataframe for pair plot
data={
    "Solar":[12000,1500,1300],
    "Wind":[3400,3600,33200],
    "Hydropower":[3400,3600,33200],
    "Biomass": [2500,2700,2400]
}

df=pd.DataFrame(data)

#Create heatmap to show the correlation matrix
correlation_matrix=df.corr()
#plot heatmap
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)

#add title
plt.title('Correlation Heatmap')
plt.show()
```
